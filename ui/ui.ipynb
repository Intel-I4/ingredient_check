```python
import cv2
from tkinter import *
from PIL import Image, ImageTk

page = 1
cap = None  # 웹캠 변수를 전역으로 선언

def next_frame():
    global page
    page += 1
    if page >= 3:
        page = 3
        frame3.lift()
        start_webcam()  # frame3이 보일 때 웹캠 시작
    elif page == 2:
        frame2.lift()
    elif page == 1:
        frame1.lift()
        stop_webcam()  # frame1에서는 웹캠 정지

def prev_frame():
    global page
    page -= 1
    if page <= 1:
        page = 1
        frame1.lift()
        stop_webcam()  # frame1에서는 웹캠 정지
    elif page == 2:
        frame2.lift()
    elif page == 3:
        frame3.lift()
        start_webcam()  # frame3이 보일 때 웹캠 시작

def load_image():
    try:
        # 이미지 로드 및 리사이징
        image = Image.open("aaaaa.jpeg")
        image = image.resize((300, 300), Image.LANCZOS)
        img = ImageTk.PhotoImage(image)
        return img
    except Exception as e:
        print(f"Error loading image: {e}")
        return None

def start_webcam():
    global cap, L1
    if cap is None:  # 웹캠이 이미 시작된 경우 다시 시작하지 않도록
        cap = cv2.VideoCapture(0)
    
        def update_camera():
              # nonlocal 키워드는 사용할 필요 없음
            if cap.isOpened():
                ret, frame = cap.read()
                if ret:
                    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                    img = ImageTk.PhotoImage(Image.fromarray(img))
                    L1['image'] = img
                    L1.image = img  # Keep a reference to avoid garbage collection
                    L1.after(10, update_camera)  # Update every 10 milliseconds
                else:
                    print("Error: Unable to read frame from webcam.")
            else:
                print("Error: Webcam not available.")
        
        update_camera()

def stop_webcam():
    global cap
    if cap is not None:
        cap.release()
        cap = None

# tkinter 윈도우 생성
root = Tk()
root.title("요리 프로젝트")
root.geometry("640x960")
root.configure(bg="#F6DDD9")

# frame3
frame3 = Frame(root, bg="#f6ddd9", border=2)
frame3.place(x=0, y=0, width=640, height=960)
Label(frame3, text="Frame3", font=('consolas', 20)).pack(fill='y')

# 카메라 로드
L1 = Label(frame3)
L1.pack()

# 이미지 로드
img = load_image()

# frame2 
frame2 = Frame(root, bg="#f6ddd9", border=2)
frame2.place(x=0, y=0, width=640, height=960)
Label(frame2, text="Frame2", bg="#f6ddd9", font=('consolas', 20)).pack(fill='y')

# frame1
frame1 = Frame(root, bg="#f6ddd9", border=2)
frame1.place(x=0, y=0, width=640, height=960)
Label(frame1, text="요린이를 위한 레시피 추천", bg="#f6ddd9", font=("Arial", 40,"bold")).place(x=110, y=200)
Label(frame1, image=img , bg="#f6ddd9").place(x=170, y=300)

# 버튼 frame1
but_frame1 = Frame(frame1, border=10, bg="#f6ddd9")
but_frame1.place(x=170, y=600, width=300, height=150)
Button(but_frame1, text='시작하기', command=next_frame, width=20, height=10, bg="#f6ddd9", font=('Helvetica', 30), border=10).pack(fill='x', expand=True)

# 버튼 frame2
but_frame2 = Frame(frame2, border=10, bg="#f6ddd9")
but_frame2.place(x=450, y=850, width=150, height=60)
Button(but_frame2, text='다음', command=next_frame).pack(side=RIGHT, padx=10)
Button(but_frame2, text='이전', command=prev_frame).pack(side=RIGHT)

# 버튼 frame3
but_frame3 = Frame(frame3, border=10, bg="#f6ddd9")
but_frame3.place(x=450, y=850, width=150, height=60)
Button(but_frame3, text='다음', command=next_frame).pack(side=RIGHT, padx=10)
Button(but_frame3, text='이전', command=prev_frame).pack(side=RIGHT)

root.mainloop()
```
